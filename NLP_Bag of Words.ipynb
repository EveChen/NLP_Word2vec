```python
def review_to_words(raw_review):
    #1. Remove html related words
    review_text = BeautifulSoup(raw_review).get_text() 
    
    #2. Remove non-letter
    letters_only = re.sub("[^a-zA-Z]", " ", review_text) 
    
    #3. Lower case & split
    words = letters_only.lower().split() 
    
    #4. Remove stop words
    stops = set(stopwords.words("english"))
    meaningful_words = [w for w in words if not w in stops]
    
    #5. Return results
    return( " ".join( meaningful_words ))
```

```python

#Test function review_to_words
junk = review_to_words(train["review"][0])
print junk
```

```python
num_reviews = train["review"].size
clean_train_reviews = []

for i in xrange(0, num_reviews):
    clean_train_reviews.append(review_to_words(train["review"][i]))
```

```python
print "Cleaning and parsing the training set movie ...\n"
clean_train_reviews = []
for i in xrange(0, num_reviews):
    if((i+1) % 1000 == 0):
        print "Review %d of %d\n" % (i+1, num_reviews)
    clean_train_reviews.append(review_to_words(train["review"][i]))
```

```python
#Show cleaning results
print clean_train_reviews[3]
print type(clean_train_reviews)   #It's a list
```

```python
print "Creating the bag of words...\n"
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = "word", \
                            tokenizer = None, \
                            preprocessor = None, \
                            stop_words = None, \
                            max_features = 5000)
print type(vectorizer)

train_data_features = vectorizer.fit_transform(clean_train_reviews)
print train_data_features[0]

train_data_features = train_data_features.toarray()
```

```python
print train_data_features
print "\n"
print train_data_features.shape
```

```python
#Now, vectorizer seems to be a model which was trained in "train_data_features
#Here, we want to see the top 5000 features from this model "vectorizer
vocab = vectorizer.get_feature_names()
print vocab
print "\n"
print vocab.__len__()  #5000
```

```python
import numpy as np  #See the counts of each word

#Previously, we got top 5000 features from "tokenizer"
#Now, we want to see the counts of each word in our train data
dist = np.sum(train_data_features, axis = 0)
print dist
print type(dist)
print dist.shape
```

```python
#To make a list - combine vocab and dist
for tag, count in zip(vocab, dist):
    print count, tag
```

```python
print train_data_features
print type(train_data_features)
```

```python
#Random Forest
print "Training the random forest..."
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 100)
forest = forest.fit(train_data_features, train["sentiment"])
```

```python
test = pd.read_csv("testData.tsv", header = 0, delimiter = "\t", quoting = 3)
print test.shape

num_reviews = len(test["review"])
clean_test_reviews = []

print "Start cleaning the data set"
for i in xrange(0, num_reviews):
    if((i+1) % 1000 == 0):
        print "Review %d of %d\n" % (i+1, num_reviews)
    clean_test_reviews.append(review_to_words(test["review"][i]))
```

```python
test_data_features = vectorizer.transform(clean_test_reviews)
test_data_features = test_data_features.toarray()

dist = np.sum(test_data_features, axis = 0)
for tag, count in zip(vocab, dist):
    print count, tag
```

```python
#Apply random forest model
result = forest.predict(test_data_features)
```

```python
#Create output
output = pd.DataFrame(data = {"id":test["id"], "sentiment":result})
print output
output.to_csv("Bag_of_words_model.csv", index = False, quoting = 3)
```
