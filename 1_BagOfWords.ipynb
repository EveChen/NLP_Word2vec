```python
#Load packages
import pandas as pd
import re
from bs4 import BeautifulSoup
import nltk
```

```python
#Read data file
train = pd.read_csv("labeledTrainData.tsv", delimiter = "\t", quoting = 3, header = 0)
print train.shape
print "\n"
print train.columns.values
print "\n"
print train["review"][0]
```

```python
#Remove html - use BeautifulSoup
removehtml = BeautifulSoup(train["review"][0])
print removehtml.get_text()
```

```python
#Remove punctuations/numbers - use re
removepun = re.sub("[^a-zA-Z]", " ", removehtml.get_text())
print removepun
```

```python
#Lower case and split
lower_case = removepun.lower()
lower_split = lower_case.split()
print lower_split
```

```python
#Remove stop words - use nltk
from nltk.corpus import stopwords
print stopwords.words("english")
removestop = [w for w in lower_split if not w in stopwords.words("english")]
print "\n"
print removestop
```

```python
#Make function to clean data including remove html, punctuations, numbers, stopwords, lower case and split

def cleandata(raw_data):
    removehtml = BeautifulSoup(raw_data).get_text()
    removepun = re.sub("[^a-zA-Z]", " ", removehtml)
    lowersplit = removepun.lower().split()
    stops = set(stopwords.words("english"))
    removestop = [w for w in lowersplit if not w in stops]
    return(" ".join(removestop))
```

```python
#Clean all train data
size = train["review"].size
cleaned_train = []
print "Start cleaning train data"

for i in xrange(0, size):
    if ((i+1) % 1000 == 0):
        print "Reviews %d in total %d\n" % (i+1, size)
    cleaned_train.append(cleandata(train["review"][i]))
```

```python
#Make vectorize & features
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = "word", stop_words = None, preprocessor = None, \
                            tokenizer = None, max_features = 3000)
train_features = vectorizer.fit_transform(cleaned_train)
train_features = train_features.toarray()
```

```python
#List the features
import numpy as np
vocab = vectorizer.get_feature_names()
print len(vocab)

dist = np.sum(train_features, axis = 0)
print dist.shape

for tag, count in zip(vocab, dist):
    print tag, count
```

```python
#Apply random forest
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 100)
forest = forest.fit(train_features, train["sentiment"])
```

```python
#Read test data
test = pd.read_csv("testData.tsv", header = 0, delimiter = "\t", quoting = 3)
print test.shape
```

```python
#Clean test data including removing html, punctuations/numbers, stop words
size = test["review"].size
cleaned_test = []

for i in xrange(0, size):
    if ((i+1) % 1000 == 0):
        print "Reviews %d of total Reviews %d\n" % (i+1, size)
    cleaned_test.append(cleandata(test["review"][i]))
```

```python
#Make features
test_features = vectorizer.transform(cleaned_test)
test_features = test_features.toarray()

dist = np.sum(test_features, axis = 0)
for count, tag in zip(vocab, dist):
    print count, tag
```

```python
#Apply the model you made to test data
result = forest.predict(test_features)
```

```python
#Create outputs
output = pd.DataFrame(data = {"id": test["id"], "sentiment":result})
output.to_csv("Practice1.csv", index = False, quoting = 3)
```
